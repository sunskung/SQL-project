### Populate Property Address data (fill in null value with existing data)

There is some null value in the property address data. But, since the parcel id will almost guarantee the name of property address. Some of the parcel ID are more than 1. Therefore, Let's use that to fill in the missing value.

````python
%%sql

SELECT propertyaddress
FROM House_Nash
WHERE propertyaddress IS NULL
LIMIT 100
````

````python
%%sql

SELECT A.parcelid, A.propertyaddress, B.parcelid, B.propertyaddress, COALESCE(A.propertyaddress, B.propertyaddress)
FROM House_Nash A
JOIN House_Nash B
ON A.parcelid = B.parcelid
AND A.uniqueid <> B.uniqueid
WHERE A.propertyaddress IS NULL
LIMIT 100
````

Let's update our data back to the original column(A)

````python
%%sql

Update House_Nash a 
SET PropertyAddress = b.PropertyAddress
FROM House_Nash b
WHERE a.ParcelID = b.ParcelID
AND a.UniqueID <> b.UniqueID 
AND a.PropertyAddress is null
RETURNING 100
````

If the table is updated, it will **not show result** on the below query which means all the null values have been filled.

````python
%%sql

SELECT A.parcelid, A.propertyaddress, B.parcelid, B.propertyaddress, COALESCE(A.propertyaddress, B.propertyaddress)
FROM House_Nash A
JOIN House_Nash B
ON A.parcelid = B.parcelid
AND A.uniqueid <> B.uniqueid
WHERE A.propertyaddress IS NULL
LIMIT 100
````

Great, all the values haas been replaced by the property data.

### Breaking out address into individual columns (spliting value in the cell)

````python
%%sql

SELECT propertyaddress
FROM House_Nash 
LIMIT 100
````

````python
%%sql

SELECT 
SUBSTRING(propertyaddress, 1, STRPOS(propertyaddress, ',')-1) AS Address,
SUBSTRING(propertyaddress, STRPOS(propertyaddress, ',')+1, LENGTH(propertyaddress)) AS Address
FROM House_Nash
LIMIT 100
````

Let's create new column to add the split values we just created.

````python
%%sql
ALTER TABLE House_Nash
ADD PropertySplitAddress VARCHAR(255)
````

````python
%%sql
UPDATE House_Nash
SET PropertySplitAddress = SUBSTRING(propertyaddress, 1, STRPOS(propertyaddress, ',')-1) 
````

### Change Y and N to yes and no in `"sold as vacant"` column

````python
%%sql

SELECT DISTINCT(soldasvacant), COUNT(soldasvacant)
FROM House_Nash
GROUP BY soldasvacant
ORDER BY count
limit 5
````

````python
%%sql
SELECT soldasvacant,
CASE WHEN soldasvacant = 'Y' THEN 'Yes'
    WHEN soldasvacant = 'N' THEN 'No'
    ELSE soldasvacant
    END
FROM House_Nash
LIMIT 100
````

Now, let's update it into out table.

````python
%%sql
UPDATE House_Nash
SET soldasvacant = CASE WHEN soldasvacant = 'Y' THEN 'Yes'
    WHEN soldasvacant = 'N' THEN 'No'
    ELSE soldasvacant
    END
````

### Remove duplicate

````python
%%sql

SELECT *,
    ROW_NUMBER() OVER(
    PARTITION BY PARCELID,
    propertyaddress,
    saleprice,
    saledate,
    legalreference
    ORDER BY uniqueid
    ) row_num
FROM House_Nash
LIMIT 500


````

With CTE, the table has the temporary column, which we can now able to call in query. Calling the row_num is now possible, and we can find out how many duplicate we have in the table with assigned columns like `PARCELID,propertyaddress,saleprice,saledate,legalreference`

````python
%%sql
WITH RowNumCTE AS(
SELECT *,
    ROW_NUMBER() OVER(
    PARTITION BY PARCELID,
    propertyaddress,
    saleprice,
    saledate,
    legalreference
    ORDER BY uniqueid
    ) row_num
FROM House_Nash
LIMIT 500
)

SELECT *
FROM RowNumCTE
WHERE row_num > 1
````

Above queries show that there are no duplicate with the assigned columns. But for the learning sake, let's write a quey to remove duplicate but not running it.

````python
%%sql
WITH RowNumCTE AS(
SELECT *,
    ROW_NUMBER() OVER(
    PARTITION BY PARCELID,
    propertyaddress,
    saleprice,
    saledate,
    legalreference
    ORDER BY uniqueid
    ) row_num
FROM House_Nash
LIMIT 500
)

DELETE
FROM RowNumCTE
WHERE row_num > 1
````

So instead of **SELECT**, we use **DELETE** query to delete the rows that has `row_num` more than 1. `row_num` that has value more than 1 means the columns `PARCELID, propertyaddress, saleprice, saledate, legalreference` has the exact same value.
